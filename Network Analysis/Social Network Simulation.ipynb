```python
import numpy as np
import pandas as pd

import networkx as nx
import random

from pylab import rcParams
import seaborn as sb
import matplotlib.pyplot as plt
```

#### Setting Notebook parameters for graphing

```python
%matplotlib inline
rcParams['figure.figsize'] = 15, 15
sb.set_style('whitegrid')
```

### Generating Dataset with Connections

#### Directional Graph with 50 Nodes

```python
node = 50
```

```python
graph = nx.gn_graph(node, seed = 777)
```

#### Edgelist

```python
edge_list = nx.generate_edgelist(graph, data=False)
```

### Assigning attributes to nodes

#### Each created node will be empty unless specifically assigned

```python
print(graph.node[0])
```

#### Populating with name attribute

```python
names = ['Ally', 'Alice', 'Andy', 'Bernard', 'Bach', 'Becky', 'California', 'Carol', 'Charles', 'Derek', 'Derol', 'Dennis',
        'Eugene', 'Eunice', 'Erin', 'Frank', 'Francis', 'Frunt', 'George', 'Gigi', 'Gordan', 'Helen', 'Henry', 'Hillary',
        'Indigo', 'Ivan', 'Ivanko', 'Jane', 'Jill', 'Janet', 'Kane', 'Kendrick', 'Kayne', 'Leonard', 'Leo', 'Lillian',
        'Maroon', 'Marcus', 'Mary', 'Noel', 'Novan', 'Norris', 'Oliver', 'Ollis', 'Oscar', 'Penny', 'Peter', 'Philip',
         'Ryan', 'Renner']
```

```python
for i in list(range(node)):
    graph.node[i]['Name'] = names[i]
```

```python
graph.node[0]
```

#### Populating with age attribute

```python
starting_age = 13
ending_age = 68
```

```python
Ages = []

for x in range (0, node):
    Ages.append(random.randint(starting_age, ending_age))
```

```python
for i in list(range(node)):
    graph.add_nodes_from([(i, {'Age': Ages[i]})])
```

```python
graph.node[0]
```

#### Populating with gender

```python
Gender = []

for x in range(0, node):
    alias = random.randint(1,2)
    
    if alias == 2:
        Gender.append('Female')
    else:
        Gender.append('Male')
```

```python
for i in list(range(node)):
    graph.node[i]['Gender'] = Gender[i]
```

```python
graph.node[0]
```

### Directional Flow Diagram

```python
nx.draw_circular(graph, node_color = 'bisque', with_labels=True)
```

#### The thick lines indicate the flow into the node. For example, taking a look at node 20, we can observe that node 37 flows into node 20!

### Visualizing the network graph

#### Adding labels to the nodes for easier visualization

```python
labeldict = {}

for i in list(range(node)):
    individual_dict = names[i]
    labeldict[i] = individual_dict
```

```python
nx.draw_circular(graph, labels=labeldict, node_color = 'bisque', with_labels=True)
```

#### Most popular person in this social network is Ally and Eugene!

### Social Network Analysis Metrics

##### 1) Degree -  describes a node's connectivity

##### 2) Successors - successor node is a node that could serve as a backup and potentially replace an influential node in a network. This serves to preserve the flow of influence throughout a network in the case where an important node is removed.

##### 3) Neighbors - adjacent nodes in a network

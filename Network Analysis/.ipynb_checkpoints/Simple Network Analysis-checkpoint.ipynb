```python
import networkx as nx
from IPython.display import Image
```

#### Initialization

```python
graph = nx.Graph()
```

```python
people = ['Harish', 'Kumar', 'Sun', 'Sara', 'Connie', 'Vinay']
```

```python
connections = [('Harish', 'Kumar'), ('Harish', 'Sun'), ('Harish', 'Vinay'), 
               ('Kumar', 'Sara'), 
               ('Sara', 'Sun'), ('Sara', 'Vinay'), 
               ('Sun', 'Vinay'), ('Sun', 'Connie')]
```

#### Adding All Nodes

```python
for node in people:
    graph.add_node(node)
```

#### Adding Edges

```python
for node_a, node_b in connections:
    graph.add_edge(node_a, node_b)
```

## Network Visualizing

```python
def get_pyplot_ax(rows = 1, columns = 1, figsize = (7, 7)):
    """ helper function """
    
    import matplotlib.pyplot as plt
    
    fig, ax = plt.subplots(rows, columns, figsize = figsize)
    
    if rows == 1 and columns == 1:
        ax.axis('off')
    elif rows == 1 or columns == 1:
        for subax in ax:
            subax.axis('off')
    elif rows > 1 or columns > 1:
        for i in range(rows):
            for j in range(columns):
                ax[i][j].axis('off')
    return fig, ax
```

### Without Colors

```python
fig, ax = get_pyplot_ax()

nx.draw_networkx(graph, ax = ax, node_size = 2300, node_color = 'white')

fig.savefig('example_friends.png', dpi = 300)
```

```python
Image(filename = 'example_friends.png', width = 500, height = 500)
```

### With Colors

#### Color coding Network

```python
ericsson_color = 'green'
external_color = 'orange'
unrelated_color = 'yellow'

color_by_company = {
    'Kumar': ericsson_color, 'Sara': ericsson_color, 'Harish': ericsson_color,
    'Sun': external_color, 'Vinay': external_color,
    'Connie': unrelated_color
}
```

#### List of Colors per node

```python
colors = [color_by_company[node] for node in graph.nodes_iter()]
```

#### Visualizing

```python
fig, ax = get_pyplot_ax()

nx.draw_networkx(graph, ax = ax, node_size = 2300, node_color = colors)

fig.savefig('example_friends_color.png', dpi = 300)
```

```python
Image(filename = 'example_friends_color.png', width = 500, height = 500)
```

## Graphical Analysis

#### Counting Number of Connections

```python
number_of_connections = graph.degree()

for connection in number_of_connections:
    print("{} has {} connections".format(connection, number_of_connections[connection]))
```

#### Spectral Clustering within Network

```python
from sklearn.cluster import SpectralClustering
```

```python
adj_mat = nx.to_numpy_matrix(graph)
```

```python
sc = SpectralClustering(3, affinity='precomputed', n_init=1000)
```

```python
sc.fit(adj_mat)
```

```python
sc.labels_
```

```python
group_colors = []
for pos, name in enumerate(graph.nodes_iter()):
    side = sc.labels_[pos]
    if side == 1:
        color = 'cyan' 
    elif side == 2:
        color = 'grey'
    else:
        color = 'red'
    group_colors.append(color)
```

```python
fig, ax = get_pyplot_ax()

nx.draw_networkx(graph, ax = ax, node_color = group_colors, node_size = 2300);

fig.savefig('example_friends_cluster.png', dpi = 300)
```

```python
Image(filename = 'example_friends_cluster.png', width = 500, height = 500)
```

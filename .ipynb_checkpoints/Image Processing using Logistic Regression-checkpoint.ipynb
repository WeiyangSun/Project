```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
%matplotlib inline
```

### Loading Digits Dataset

```python
digits = load_digits()
```

#### Determining total number of images and labels

```python
print('Image Data Shape: ', digits.data.shape)
```

```python
print('Label Data Shape: ', digits.target.shape)
```

### Taking a look at some images from Dataset

```python
plt.figure (figsize = (20,4))

i = 5
for index, (image, label) in enumerate(zip(digits.data[0:i], digits.target[0:i])):
    plt.subplot(1, i, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap = plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize = 20)
```

### Splitting Dataset for Train-Test

```python
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.33, random_state = 777)
```

```python
print('Train dataset shape: ', x_train.shape)
```

```python
print('Test dataset shape: ', x_test.shape)
```

### Deploying Logistic Regression Model

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
predictions = logreg.predict(x_test)
```

```python
score = logreg.score(x_test, y_test)
score
```

### Representing confusion matrix in heat map

```python
cnf_matrix = confusion_matrix(y_test, predictions)
```

```python
plt.figure(figsize = (20,20))

sns.heatmap(cnf_matrix, annot = True, fmt = '.2f', linewidths = .5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual Labels');
plt.xlabel('Predicted Labels');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 20);
```

```python
index = 0
classified_list = []

for predict, actual in zip(predictions, y_test):
    
    if predict == actual:
        classified_list.append(index)
        
    index += 1
```

```python
plt.figure(figsize = (20,20))

for plotIndex, wrong in enumerate(classified_list[0:3]):
    
    plt.subplot(1,3, plotIndex + 1)
    plt.imshow(np.reshape(x_test[wrong], (8,8)), cmap = plt.cm.gray)
    plt.title('Predicted: {}, Actual: {}'.format(predictions[wrong], y_test[wrong]), fontsize = 20)
```
